# GGPlot in python Demo 

---

**Importing libs**

```python
import pandas as pd
from plotnine import *
```

**Loading dataset from URL on GitHub**

```python
url = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/insurance.csv"
df = pd.read_csv(url)
df.describe()
```

```python
df.head()
```

**Create list of region**

```python
region = ['souteast','southwest','northeast','northwet']
```

**Box-Plot**

```python
from plotnine.scales.scale_xy import scale_x_discrete
(
    ggplot(df)
    + geom_boxplot(aes(x='factor(region)',y='bmi',color='region'))
    + scale_x_discrete(labels=region, name=region)
)
```

**Scatter-Plot**

```python
(
    ggplot(df,aes(x='charges',y='bmi',color='smoker'))
    +geom_point()
    +facet_grid('children ~ .')
    +labs(x='charges',y='bmi')
    +theme_538()
)
```

**Subplot of histograms**

```python
(
    ggplot(df) 
    + geom_histogram(mapping = aes(x='charges', color='smoker'), bins=20, fill='white')
    + facet_grid('sex  ~ region')
)
```

**Buuble chart of counts**

```python
(
    ggplot(df) 
    + geom_count(mapping = aes(x='region', y='sex', color='region')) 
    + labs(title = "Number of Observations in Each Region")
)
```

**Histogram**

```python
(
    ggplot(df) 
    + geom_histogram(mapping = aes(x='charges', color='smoker'), bins = 15)
)
```
